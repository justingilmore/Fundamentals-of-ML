```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 100);
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')

plt.legend();
```

 To set x-lim, x-label, and title properties of the axes, use the _set()_ method

```python
ax = plt.axes()
x = np.linspace(0, 10, 100);

ax.plot(x, np.sin(x))
ax.plot(x, np.cos(x))
ax.set(xlim=(0, 10),
       xlabel='x',
       title='Sinusoids: sin(x) and cos(x)');
```

Similarly, for only axes titles, use the _set()_ method

```python
ax = plt.axes()
x = np.linspace(0, 10, 100);

ax.plot(x, np.sin(x))
ax.plot(x, np.cos(x))
ax.set(xlabel='x', ylabel='y');
```

The second central moment of a random variable is its variance: $Var(X) = E(X^2)-(E(X))^2$, where $E(X)$ is the Expected value or mean of the random variable X.  The covariance of X and Y is $Cov(X,Y)= E(XY)-\mu_x\mu_y$, or the correlation minus the product of the means.  This can be thought of as the central moment of the two random variables.  If X and Y are independent, $E(XY) = E(X)E(Y)=\mu_x\mu_y$ and $Cov(X,Y)=0$. Then, the two random variables are uncorrelated.
